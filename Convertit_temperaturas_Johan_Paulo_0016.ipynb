Fórmulas de ConversiónPara convertir entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:De Celsius a Fahrenheit:$$°F = (°C \times \frac{9}{5}) + 32$$(O de forma simplificada: $°F = (°C \times 1.8) + 32$)De Fahrenheit a Celsius:$$°C = \frac{(°F - 32) \times 5}{9}$$(O de forma simplificada: $°C = \frac{°F - 32}{1.8}$)2. Ejemplos Resueltos Paso a PasoCaso A: Convertir 25°C a Fahrenheit (Temperatura ambiente ideal)Identificar datos: Tenemos $25°C$.Sustituir en la fórmula: $°F = (25 \times 1.8) + 32$.Multiplicar: $25 \times 1.8 = 45$.Sumar: $45 + 32 = 77$.Resultado: 77°F.Caso B: Convertir 98.6°F a Celsius (Temperatura corporal promedio)Identificar datos: Tenemos $98.6°F$.Sustituir en la fórmula: $°C = \frac{98.6 - 32}{1.8}$.Restar primero: $98.6 - 32 = 66.6$.Dividir: $66.6 / 1.8 = 37$.Resultado: 37°C.3.
